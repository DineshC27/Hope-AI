```python
import pandas as pd
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
datasets = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
datasets
```

```python
datasets.columns
```

```python
independent = datasets[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent = datasets[['Profit']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
xtrain
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='squared_error',splitter='random',max_features='log2')
regressor = regressor.fit(xtrain,ytrain)
```

```python
#to view the tree structure of this dataset
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
ypredict = regressor.predict(xtest)
```

```python
from sklearn.metrics import r2_score
rscore = r2_score(ytest,ypredict)
```
