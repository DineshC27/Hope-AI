```python
# In this project pandas class library is used to invoke the IO operations
import pandas
dataset = pandas.read_csv("Salary_Data.csv") #loading the salary detail from the system
```

```python
dataset #printing the value in this variable(Dataset)
```

```python
#Fetch only the year of experience value from the dataset and assign it to a variable called independent
independent = dataset[["YearsExperience"]] 
independent #Print the value
```

```python
#Fetch only the salary value from the dataset and assign it to a variable called dependent
dependent=dataset[["Salary"]]
dependent #Print the value
```

```python
#from the sklearn.model_selection library we import only train_test_split class to split the train and test data.
#This kind of import load only the perticular class which save the memory
from sklearn.model_selection import train_test_split
```

```python
#Splitting the training and test data and assigning it to the corresponding variables
#Always 1st 2 variables has train & test input values, 2nd 2 variables has train & test output values.
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,train_size=0.70,random_state=0)
```

```python
x_train # Just verfying the data by printing it
```

```python
#from the sklearn.linear_model library we import only LinearRegression class to train our model.
from sklearn.linear_model import LinearRegression
regression = LinearRegression() # Assigning all the functions in LinearRegression() class to regression variable.
regression.fit(x_train,y_train) #Train our model by passing the Input & output training data
```

```python
#Just view the value of weight and bias calculated in Simple linear Regression
#y=wx+b (or) Output = weight * input + bias
weight=regression.coef_
bias=regression.intercept_
weight
```

```python
bias
```

```python
# After our model is ready, predict the output values for all the input test data
y_predict = regression.predict(x_test) 
```

```python
y_predict # Just verfying the data by printing it
```

```python
#From the sklearn.metrics library we import only r2_score class to test our model.
from sklearn.metrics import r2_score
```

```python
#Passing the sample output value and model predicted output value as a parameter 
#to calculate the r square value.
rscore = r2_score(y_test,y_predict)
```

```python
# Verfying the r squaer value, if the value is nearly 1 means our model is performing well otherwise it is a not
rscore 
```

```python
#Invoking the class pickle to save and reloade our model
import pickle
```

```python
#Save our model to the local system
#wb indicate write binary
pickle.dump(regression,open("SlaryPredict.sav",'wb'))
```

```python
#Load our model from the local system to validate it
ai_model=pickle.load(open("SlaryPredict.sav",'rb'))
```
