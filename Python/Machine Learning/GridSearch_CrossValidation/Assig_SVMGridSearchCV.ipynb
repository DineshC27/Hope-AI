```python
import pandas as pd
dataset = pd.read_csv("Sample data.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent = sc.fit_transform(independent)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
param = {'kernel':['linear','poly','rbf','sigmoid'],'C':[1,10,100],'gamma':['auto','scale']}
model = GridSearchCV(SVR(),param,refit = True, verbose = 3,n_jobs=-1)
```

```python
model.fit(independent,dependent.values.ravel())
```

```python
re = model.cv_results_
```

```python
table = pd.DataFrame.from_dict(re)
```

```python
table
```

```python
model.best_params_
```
