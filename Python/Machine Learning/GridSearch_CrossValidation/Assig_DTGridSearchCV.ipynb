```python
import pandas as pd
dataset = pd.read_csv("Sample data.csv")
```

```python
dataset
```

```python
datasets = pd.get_dummies(dataset,dtype=int,drop_first=True)
datasets
```

```python
datasets.columns
```

```python
independent = datasets[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent = datasets[['charges']]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'max_features': ['sqrt','log2'],'splitter':['best','random']}
```

```python
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1)

grid.fit(independent, dependent)
```

```python
re=grid.cv_results_
```

```python
grid.best_params_
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```
