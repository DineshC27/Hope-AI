```python
import pandas as pd
dataset = pd.read_csv("Sample data.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
dependent = dataset[['charges']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param = {'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features':['sqrt','log2', None],'n_estimators':[10,100]}
```

```python
model = GridSearchCV(RandomForestRegressor(),param,refit = True, verbose = 3,n_jobs=-1)
```

```python
model.fit(independent, dependent.values.ravel())
```

```python
re = model.cv_results_
```

```python
model.best_params_
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
import pickle
pickle.dump(model,open('RFGridSearchCV.sav','wb'))
```
