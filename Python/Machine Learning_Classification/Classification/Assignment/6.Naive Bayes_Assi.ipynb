```python
import pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
indep = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep = dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
# This block is for GaussianNB in Naive Bayes
from sklearn.naive_bayes import GaussianNB

param = {}
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(GaussianNB(),param,refit=True,verbose=3,scoring='f1_weighted',n_jobs=-1,cv=5)

model.fit(xtrain,ytrain.values.ravel())
ypredict = model.predict(xtest)
model.best_params_
re = model.cv_results_
table = pd.DataFrame.from_dict(re)
table
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypredict)
from sklearn.metrics import classification_report
clfrpt = classification_report(ytest,ypredict)
print (clfrpt)
print(cm)
from sklearn.metrics import f1_score
f1 = f1_score(ytest,ypredict,average='weighted')
print (f1)
from sklearn.metrics import roc_auc_score
rocauc = roc_auc_score(ytest,model.predict_proba(xtest)[:,1])
print (rocauc)
```
