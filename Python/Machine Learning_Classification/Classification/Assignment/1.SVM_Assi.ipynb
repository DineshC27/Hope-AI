```python
import pandas as pd
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset['classification_yes'].value_counts()
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
from sklearn.svm import SVC
```

```python
#param = {'kernel':['linear', 'poly','rbf','sigmoid', 'precomputed'],
#         'gamma':['scale','auto'],'C':[10,100,1000,2000,3000]}
param = {'kernel':['linear', 'poly','rbf','sigmoid'],
         'gamma':['scale','auto'],'C':[10,100,1000,2000,3000]}
```

```python
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(SVC(probability=True),param,refit=True,verbose=3,scoring='f1_weighted',n_jobs=-1,cv=5)
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
ypredict = model.predict(xtest)
```

```python
model.best_params_
```

```python
re = model.cv_results_
table = pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ytest,ypredict)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
rs = classification_report(ytest,ypredict)
```

```python
print(rs)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(ytest,ypredict,average='weighted')
print (f1)
```

```python
from sklearn.metrics import roc_auc_score
rocauc = roc_auc_score(ytest,model.predict_proba(xtest)[:,1])
print (rocauc)
```
