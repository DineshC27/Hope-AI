```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
x
```

```python
#Below are not part of this algorithm
import os
os.environ['LOKY_MAX_CPU_COUNT'] = '2'  
```

```python
#below code is to find cluster value
from sklearn.cluster import SpectralClustering
#spec = SpectralClustering(n_clusters=4,eigen_tol='auto', assign_labels='kmeans')
#{'polynomial', 'additive_chi2', 'chi2', 'linear', 'precomputed_nearest_neighbors', 'laplacian', 'sigmoid', 'precomputed', 'poly', 'nearest_neighbors', 'cosine', 'rbf'}
spec = SpectralClustering(n_clusters= 4, assign_labels='discretize',random_state=3,affinity='poly')
#spec.fit(x)
ypred=spec.fit_predict(x)
ypred
```

```python
supervised = dataset
supervised['Cluster_Group'] = ypred
```

```python
import seaborn as sns
facets = sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True)
```
