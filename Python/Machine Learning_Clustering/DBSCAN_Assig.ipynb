```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
x=dataset.iloc[:,[3,4]]
x
```

```python
from sklearn.cluster import DBSCAN
#clumodel = DBSCAN(eps=0.5, min_samples=5, metric='euclidean',algorithm='brute')
clumodel = DBSCAN(eps=13, min_samples=1, metric='nan_euclidean',algorithm='brute')
ypred=clumodel.fit_predict(x)
ypred
```

```python
supervised = dataset
supervised['Cluster_Group']=ypred
```

```python
import seaborn as sns
facets = sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True)
```
